```python
import matplotlib.pyplot as plt
import operator
```

```python
fname = open("assignment2.txt","r")
d1 = dict()
count = 0
for line in fname:
    words = line.split()
    count += len(words)
    for word in words:
        if word in d1:
            d1[word] = d1[word]+1
        else:
            d1[word] = 1
print("Count of words is: ",count) 
```

```python
sorted_d=(sorted(d1.items(), key =operator.itemgetter(1), reverse=True))[::1]

x_val = [ x[0] for x in sorted_d]
y_val = [ x[1] for x in sorted_d] 



```

```python
fig, ax = plt.subplots(figsize =(70,20))
#print(d1.keys())
label = x_val[0]  # highest frequency word

print("Word with highest frequency is:",label)

plt.bar(x_val, y_val,color='g')

plt.bar(x_val, y_val,color='g')

plt.xticks(rotation=90, fontsize=18)
plt.title("A Word frequency plot for 'assignment2.txt' file")
plt.ylabel("Frequency ---->")
plt.xlabel("Words---->")
plt.show() 
```
